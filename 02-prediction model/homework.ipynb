# this kind of distribution is not good for mahcine learning, so we want to get rid of long tail.So we apply log to get compact values

````python
np.log([1,10,1000,100000])
````

````python
# complains with zerooo
np.log([0,1,10,1000,100000])
````

````python
# we add one to everything lol
np.log([0+1,1+1,10+1,1000+1,100000+1])
````

````python
# sb smart made a shortcut for this lol agaIN
np.log1p([0,1,10,1000,100000])
````

````python
price_logs = np.log1p(df.msrp)
````

````python
sns.histplot(price_logs,bins =50) #as you can see the tails are gone
# and this looks more/resembles a normal distro
````
