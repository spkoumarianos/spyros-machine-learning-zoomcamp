```python
u=np.array([2,4,5,6])
v=np.array([1,0,0,2])
```

```python
#vector vector multiplication dot prod or inner prod

def vector_vector_multiplication(a, b):
    assert a.shape[0] == b.shape[0] #same size and the zero is for just the number not a dtype
    n=b.shape[0]
    result = 0.0
    for i in range(n):
        result = result + a[i]*b[i]

    return result
```

```python
#test this 
vector_vector_multiplication(u, v)
```

```python
# this already exists in numpy so we can directly use 
u.dot(v)
```

```python
U = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1]
])
```

```python
v
```

```python
# vectro mtrx mult
def matrix_vector_multiplication(A,b):
    assert A.shape[1] == b.shape[0]
    num_rows = A.shape[0]
    result = np.zeros(num_rows)
    for i in range(num_rows):
        result[i] = vector_vector_multiplication(A[i],b)
    return result    
```

```python
matrix_vector_multiplication(U,v)
```

```python
U.dot(v)
```

```python
V=np.array([
    [1, 1, 2],
    [0, 0.5, 1],
    [0, 2, 1],
    [2, 1, 0]
])
```

```python
# mtrx mtrx mult
def matrix_matrix_multiplication(A,B):
    assert A.shape[1] == B.shape[0]
    num_rows = A.shape[0]
    num_cols = B.shape[1]
    result = np.zeros((num_rows, num_cols))
    for i in range(num_cols):
        bi=B[:, i]
        Abi = matrix_vector_multiplication(A,bi)
        result[:, i] = Abi
    return result    
```

```python
matrix_matrix_multiplication(U,V)
```

```python
U.dot(V)
```
